```python
# 감시 피하기
'''
선생님T 학생S 장애물O

학생이 선생님의 시야를 다 벗어날 수 있는지

좌표 기준은 (0,0) ... (n-1,n-1) 로 하자.
'''

from itertools import combinations
import copy

n = int(input())

maps = []
for i in range(n):
    maps.append(input().split())


maps2 = copy.deepcopy(maps)

# 선생님의 감시를 수행하는 함수
# 선생님 좌표를 입력받고
# 걸리는 학생이 있으면 True 없으면 False
def sight(x, y):
    global n, maps
    
    if x<0 or y<0 or x>=n or y>=n:
        return
    
    tx = x
    ty = y
    
    while(True):
        tx+=1
        if tx>=n:
            break
        if maps[tx][ty] == 'O':
            break
        elif maps[tx][ty] == 'S':
            return True
    
    tx = x
    ty = y
    
    while(True):
        tx-=1
        if tx<0:
            break
        if maps[tx][ty] == 'O':
            break
        elif maps[tx][ty] == 'S':
            return True
    
    tx = x
    ty = y
    
    while(True):
        ty+=1
        if ty>=n:
            break
        if maps[tx][ty] == 'O':
            break
        elif maps[tx][ty] == 'S':
            return True
    
    tx = x
    ty = y
    
    while(True):
        ty-=1
        if ty<0:
            break
        if maps[tx][ty] == 'O':
            break
        elif maps[tx][ty] == 'S':
            return True
    
    return False


'''
X 중 3곳을 O로 만들기
반복해서 S의 위치마다 sight함수 호출
한번이라도 True 나오면 패스하고 다음 차례 체크

반복 마쳤을 때까지 다 False가 있다 -> YES 출력 후 프로그램 종료
'''

# O 3곳 만들기 (반복문)
place = []
for i in range(n):
    for j in range(n):
        if maps[i][j]=='X':
            place.append((i,j))

printYes = False
for a in combinations(place, 3):
    for i in range(3):
        maps[a[i][0]][a[i][1]] = 'O'
    # S 위치마다 sight 호출
    check = False
    for i in range(n):
        for j in range(n):
            if maps[i][j] == 'T':
                if sight(i, j) == True:
                    check = True
                    break
        if check == True:
            break
    
    if check == False:
        print(maps)
        print('YES')
        printYes = True
        break
    maps = copy.deepcopy(maps2)

if printYes == False:
    print('NO')
```
