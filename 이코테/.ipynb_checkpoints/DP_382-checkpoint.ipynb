```python
# 편집 거리
'''
길이가 같다면 -> replace or 유지
if len(a) == len(b):
    for i:
        if b[i]!=a[i]:
            replace
길이가 다르다면 -> 앞에부터 비교 -> 같으면 pass 다르면 같은걸로 insert
while len(a) != len(b):
    for i:
        if b[i]!=a[i]:
            insert (길이 같게 만들기)
'''

before = list(input())
after = list(input())

lb = len(before)
la = len(after)

if lb>la:
    # remove
    i = -1
    while(True):
        i+=1
        if before[i] == after[i]:
            continue
        else:
            before.pop(i)
        if len(before) == len(after):
            break

elif lb<la:
    # insert
    i = -1
    while(True):
        print(i)
        i+=1
        if before[i] == after[i]:
            continue
        else:
            before.insert(i, after[i])
        if len(before) == len(after):
            break
print(before)
print(after)

'''
이거 아닌거같은데
abcdefghi
iabcdefgh
'''
```

```python
# 답안지

def edit_dist(str1, str2):
    n = len(str1)
    m = len(str2)
    
    dp = [[0] * (m+1) for _ in range(n+1)]
    
    for i in range(n+1):
        dp[i][0] = i
    for j in range(m+1):
        dp[0][j] = j

    for i in range(1, n+1):
        for j in range(1, m+1):
            if str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            
            else:
                dp[i][j] = 1 + min(dp[i][j-1], dp[i-1][j], dp[i-1][j-1])
    
    return dp

str1 = 'abcdefghijklmnopqrstuvwxyz'
str2 = 'zabcdefghijklmnopqrstuvwxy'
answer = edit_dist(str1, str2)
for i in answer:
    print(i)
```
